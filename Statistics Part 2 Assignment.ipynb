# **Statistics Part 2 Assignment**

1. **What is hypothesis testing in statistics?**
    - Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves testing an assumption (hypothesis) to determine whether there is enough evidence to support it.

2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**
    - The null hypothesis (H₀) assumes no effect or no difference. The alternative hypothesis (H₁) proposes that there is an effect or a difference. Hypothesis testing evaluates whether to reject H₀ in favor of H₁.

3. **What is the significance level in hypothesis testing, and why is it important?**
    - The significance level (α) is the threshold for rejecting the null hypothesis. Commonly set at 0.05 or 5%, it indicates the probability of making a Type I error (false positive).



4. **What does a P-value represent in hypothesis testing?**
     - The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis is true. A small P-value (< α) suggests strong evidence against H₀.

5. **How do you interpret the P-value in hypothesis testing?**
    - If P-value ≤ α, reject the null hypothesis (evidence supports H₁). If P-value > α, do not reject H₀ (insufficient evidence to support H₁).

6. **What are Type 1 and Type 2 errors in hypothesis testing?**
    - Type 1 Error: Rejecting a true null hypothesis (false positive).

   - Type 2 Error: Failing to reject a false null hypothesis (false negative).

7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
    - One-tailed test checks for a deviation in one direction (greater or less).

    - Two-tailed test checks for deviations in both directions (not equal).

8. **What is the Z-test, and when is it used in hypothesis testing?**
    - A Z-test is used to test hypotheses when the population variance is known and the sample size is large (n > 30). It compares sample mean to population mean.

9. **How do you calculate the Z-score, and what does it represent in hypothesis testing?**
    - Z-score = (X̄ - μ) / (σ / √n), where X̄ is sample mean, μ is population mean, and σ is standard deviation. It shows how many standard deviations a data point is from the mean.



10. **What is the T-distribution, and when should it be used instead of the normal distribution?**
    - The T-distribution is used when the sample size is small (n < 30) and population standard deviation is unknown. It's wider than the normal distribution.



11. **What is the difference between a Z-test and a T-test?**
    - Z-test: Known population variance, large sample.

    - T-test: Unknown population variance, small sample.

12. **What is the T-test, and how is it used in hypothesis testing?**
    - The T-test compares sample means to determine if there is a statistically significant difference between them, especially when population variance is unknown.

13. **What is the relationship between Z-test and T-test in hypothesis testing?**
    - Both are used to test means, but T-tests are used for small samples or unknown variance, while Z-tests are used when variance is known and sample is large.

14. **What is a confidence interval, and how is it used to interpret statistical results?**
    - A confidence interval provides a range of values within which the true population parameter is likely to lie. A 95% CI means there's 95% confidence the interval contains the true value.

15. **What is the margin of error, and how does it affect the confidence interval?**
    - The margin of error is the maximum expected difference between the observed statistic and the true parameter. A larger margin creates a wider confidence interval.



16. **How is Bayes' Theorem used in statistics, and what is its significance?**
    - Bayes' Theorem calculates conditional probabilities. It's significant in updating probabilities based on new evidence, commonly used in decision making and AI.



17. **What is the Chi-square distribution, and when is it used?**
    - Chi-square distribution is used for categorical data analysis. It's applied in tests of independence and goodness-of-fit tests.

18. **What is the Chi-square goodness of fit test, and how is it applied?**
    - This test determines if a sample matches a population with a specified distribution. It compares observed vs. expected frequencies.

19. **What is the F-distribution, and when is it used in hypothesis testing?**
    - F-distribution is used to compare two variances. It is applied in ANOVA and tests involving variance analysis.

20. **What is an ANOVA test, and what are its assumptions?**
    - ANOVA (Analysis of Variance) compares means of 3 or more groups. Assumptions: independence, normality, and equal variance across groups.

21. **What are the different types of ANOVA tests?**
    - One-Way ANOVA: One independent variable.

    - Two-Way ANOVA: Two independent variables.

    - Repeated Measures ANOVA: Same subjects tested multiple times.

22. **What is the F-test, and how does it relate to hypothesis testing?**
    - The F-test compares variances between groups. In hypothesis testing, it helps determine if the variability between group means is larger than expected by chance.

1. **Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.**
   - To perform a Z-test in Python, you can use the statsmodels or scipy library. The Z-test checks if a sample mean significantly differs from a known population mean.

```python
import numpy as np
from scipy import stats

sample = np.array([88, 92, 85, 91, 89])
population_mean = 90
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 1 - stats.norm.cdf(abs(z_score))

print("Z-Score:", z_score)
print("P-Value:", p_value)
```

2. **Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.**
    - You can simulate data using numpy and perform tests using scipy.stats.

```python
import numpy as np
from scipy import stats

data = np.random.normal(loc=100, scale=15, size=30)
t_stat, p_value = stats.ttest_1samp(data, popmean=105)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

3. **Implement a one-sample Z-test using Python to compare the sample mean with the population mean.**

```python
def one_sample_z_test(sample_data, population_mean):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    n = len(sample_data)
    z = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p

sample = np.array([105, 102, 98, 100, 101])
z, p = one_sample_z_test(sample, 100)
print("Z-Score:", z, "P-value:", p)
```

4. **Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import matplotlib.pyplot as plt

z_score = 2.1
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -1.96) | (x >= 1.96), color='red', alpha=0.3)
plt.axvline(z_score, color='blue', linestyle='--')
plt.title("Two-Tailed Z-Test with Critical Region")
plt.show()
```

5. **Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
def visualize_type_errors():
    x = np.linspace(-4, 4, 1000)
    y1 = stats.norm.pdf(x, 0, 1)  # H0
    y2 = stats.norm.pdf(x, 1.5, 1)  # H1

    plt.plot(x, y1, label="H0", color='blue')
    plt.plot(x, y2, label="H1", color='green')
    plt.fill_between(x, y1, where=(x > 1.96), color='red', alpha=0.3, label="Type I Error")
    plt.fill_between(x, y2, where=(x < 1.96), color='orange', alpha=0.3, label="Type II Error")
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

visualize_type_errors()
```

6. **Write a Python program to perform an independent T-test and interpret the results.**

```python
group1 = np.random.normal(100, 10, 30)
group2 = np.random.normal(105, 10, 30)

t_stat, p_val = stats.ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_val)
```

7. **Perform a paired sample T-test using Python and visualize the comparison results.**

```python
before = np.random.normal(100, 5, 20)
after = before + np.random.normal(2, 3, 20)

t_stat, p_val = stats.ttest_rel(before, after)
print("Paired T-test T-statistic:", t_stat)
print("P-value:", p_val)

plt.plot(before, label="Before")
plt.plot(after, label="After")
plt.legend()
plt.title("Before vs After - Paired Samples")
plt.show()
```

8. **Simulate data and perform both Z-test and T-test, then compare the results using Python.**

9. **Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    std_err = stats.sem(data)
    margin = stats.t.ppf((1 + confidence) / 2, len(data)-1) * std_err
    return mean - margin, mean + margin

sample = np.random.normal(100, 10, 30)
ci = confidence_interval(sample)
print("95% Confidence Interval:", ci)
```
